```python
# Import Libaries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Datasets
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_original = df.copy()
df_original['job_posted_month'] = df['job_posted_date'].dt.strftime('%b')
```

```python
oct_df = df_original[df_original['job_posted_month'] == 'Oct'].copy()

nov_df = df_original[df_original['job_posted_month'] == 'Nov'].copy()

dec_df = df_original[df_original['job_posted_month'] == 'Dec'].copy()


```

```python
df_q4 = pd.concat([oct_df, nov_df, dec_df], ignore_index=True)
```

```python
df_q4['job_posted_month'].value_counts().plot(kind='bar')
```
