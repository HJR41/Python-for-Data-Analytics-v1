```python
# Import Libaries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Datasets
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
#Data frame with just Data Analyst Jobs
df_DA = df[df['job_title_short'] == 'Data Analyst'].copy()
```

```python


df_DA['job_posted_month_no'] = df_DA['job_posted_date'].dt.month

```

```python
df_DA_explode = df_DA.explode('job_skills')
```

```python
df_DA_pivot = df_DA_explode.pivot_table(index='job_posted_month_no', columns='job_skills', aggfunc='size', fill_value=0) #aggfunc ='size' counts the number of times each skills is posted in each month (the index)
```

```python
#.loc Total adds a row called total - .sum puts the sum of all the skills in the total row

df_DA_pivot.loc['Total'] = df_DA_pivot.sum()

#Sort the columns by their totals.Then add it to the df to provide the pivot ordered by totals 

df_DA_pivot = df_DA_pivot[df_DA_pivot.loc['Total'].sort_values(ascending=False).index]
```

```python
#drop totals column:

df_DA_pivot = df_DA_pivot.drop('Total')

df_DA_pivot
```

```python
df_DA_pivot.iloc[:, :5].plot(kind='line')

plt.title('Count per month of the Top 5 Data Analyst Skills')
plt.ylabel('Mention Count')
plt.xlabel('Month')
plt.show()
```
