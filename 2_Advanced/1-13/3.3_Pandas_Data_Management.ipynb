```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_original = df.copy()
```

```python
df_salary_nona = df_original.copy()
```

```python
median_salary_year = df_salary_nona['salary_year_avg'].median()
```

```python
df_salary_nona['salary_year_avg'] = df_salary_nona['salary_year_avg'].fillna(median_salary_year)
```
