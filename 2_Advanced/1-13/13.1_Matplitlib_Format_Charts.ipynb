```python
# Import Libaries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Datasets
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_location = df.copy()
```

```python
df_postings = df.copy()
```

```python
df_location_top3 = df_location['job_location'].value_counts().head(3)
df_location_top3
```

```python
df_postings_top3 = df_postings['job_title_short'].value_counts().head(3)
```

```python
fig, ax = plt.subplots(2, 1)

df_location_top3.plot(kind='bar', ax=ax[0])
df_postings_top3.plot(kind='barh', ax=ax[1])
ax[1].invert_yaxis()

fig.tight_layout()
```
