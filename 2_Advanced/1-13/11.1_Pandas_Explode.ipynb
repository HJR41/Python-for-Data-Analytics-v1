```python
# Import Libaries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Datasets
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python

df_cloud_eng = df.copy()
```

```python
df_cloud_eng = df_cloud_eng[df_cloud_eng['job_title_short'] == 'Cloud Engineer']
```

```python
df_cloud_exploded = df_cloud_eng.explode('job_skills')

df_cloud_exploded['job_skills'].value_counts().head(10).plot(kind='barh')
```
