```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_original = df.copy()
```

```python
df_original.pivot_table(values='salary_year_avg', index='job_country',columns='job_title_short', aggfunc='median')
#long tables > Wide Tables
```

```python
#find the top 6 countries by posting count
top_countries = df_original['job_country'].value_counts().head(6).index

top_countries
```

```python
df_job_country_salary = df_original.pivot_table(
    values='salary_year_avg',
    index='job_country',
    columns='job_title_short',
    aggfunc='median'
    )

df_job_country_salary = df_job_country_salary.loc[top_countries]

job_titles = ['Data Analyst', 'Data Engineer', 'Data Scientist']

df_job_country_salary = df_job_country_salary[job_titles]

df_job_country_salary.plot(kind='bar')
plt.ylabel('Median Salary (USD)')
plt.xlabel('Country')
plt.title('Median Salary by Country')
plt.xticks(rotation=45, ha='right')
plt.show()
```
