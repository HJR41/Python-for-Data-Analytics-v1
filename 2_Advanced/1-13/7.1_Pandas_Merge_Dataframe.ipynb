```python
# Import Libaries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Datasets
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_original = df.copy()#
```

```python
df_original['job_posted_month'] = df_original['job_posted_date'].dt.to_period('M')
df_us_jobs = df_original[df_original['job_country'] == 'United States'].copy()
```

```python
df_monthly_postings = df_original.groupby('job_posted_month').size().reset_index(name='postings_count')
```

```python
merged_df = pd.merge(df_us_jobs, df_monthly_postings, on='job_posted_month', how='right')
merged_df.head()
```
