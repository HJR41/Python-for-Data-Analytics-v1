```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_original = df.copy()
```

```python
df_company_salary = df_original.copy()
```

```python
df_company_salary = df_company_salary.pivot_table(
    values='salary_year_avg',
    index='company_name',
    aggfunc=['max', 'min', 'mean']
)

```

```python
df_company_salary = df_company_salary.head()
df_company_salary
```

```python
df_company_salary.plot(kind='bar')
plt.ylabel('Yearly Salary (USD)')
plt.xlabel('')
plt.xticks(rotation=45, ha='right')
plt.title('Max, Min & Mean Salary per Company')
plt.legend(title='')
```
