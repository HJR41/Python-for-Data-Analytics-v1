```python
# Import Libaries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Datasets
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_original = df.copy()
df_original
```

```python
# 'job_schedule_type'

full_time_df = df_original[df_original['job_schedule_type'] == 'Full-time']

part_time_df = df_original[df_original['job_schedule_type'] == 'Part-time']
```

```python
pt_ft_combined_df = pd.concat([full_time_df, part_time_df], ignore_index=True)
```

```python
pt_ft_combined_df['job_schedule_type'].value_counts().plot(kind='bar')
```
