```python
# Import Libaries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Datasets
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_original = df.copy()

```

```python
data_scientist_df = df_original[df_original['job_title'] == 'Data Scientist'].copy()

data_analyst_df = df_original[df_original['job_title'] == 'Data Analyst'].copy()
```

```python
combined_df = pd.concat([data_analyst_df, data_scientist_df], ignore_index=True)
```

```python
combined_df['job_title'].value_counts().plot(kind='bar')
```
