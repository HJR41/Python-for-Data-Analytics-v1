```python
# Import Libaries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Datasets
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_ft_pt = df[df['job_schedule_type'].isin(['Full-time', 'Part-time'])].copy()
plt_ft_pt = df_ft_pt['job_schedule_type'].value_counts()
```

```python

plt.pie(plt_ft_pt, labels=plt_ft_pt.index, autopct='%1.1f%%', startangle=90)
plt.title('Job Schedule Full-time vs Part-time')
plt.show()
```
