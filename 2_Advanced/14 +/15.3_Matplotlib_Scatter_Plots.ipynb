```python
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Datasets
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
filtered_df = df[
    (df['job_title_short'].isin(['Data Analyst', 'Data Scientist', 'Data Engineer'])) &
    (df['job_country'] == 'United States') &
    (df['job_schedule_type'].isin(['Full-time', 'Contractor'])) &
    (df['salary_year_avg'].notna())
].copy()

# Creating the job_title_type column
filtered_df['job_title_type'] = filtered_df['job_title_short'] + ' - ' + filtered_df['job_schedule_type']

# Grouping by job_title_type and calculating averages
grouped_df = filtered_df.groupby('job_title_type').agg(
    avg_salary=('salary_year_avg', 'mean'),
    job_count=('job_title_type', 'count')
).reset_index()

# Plotting the scatter plot
plt.figure(figsize=(10, 6))
for job_title_type in grouped_df['job_title_type'].unique():
    subset = grouped_df[grouped_df['job_title_type'] == job_title_type]
    plt.scatter(subset['job_count'], subset['avg_salary'], label=job_title_type)

# Adding labels and title
plt.xlabel('Job Count')
plt.ylabel('Average Yearly Salary (USD)')
plt.title('Salary vs Job Count for Data Roles in the US')
plt.legend()
plt.grid(True)
plt.show()
```
