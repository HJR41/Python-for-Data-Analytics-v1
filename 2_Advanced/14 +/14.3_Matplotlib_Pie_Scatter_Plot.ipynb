```python
# Import Libaries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Datasets
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_us = df[df['job_country'] == 'United States'].copy()
```

```python
degree_mentioned = df_us['job_no_degree_mention'].value_counts()
```

```python
explode = [0, 0.1]
plt.pie(degree_mentioned, explode=explode, labels=['Degree Required','No Degree Required'], autopct='%1.1f%%', startangle=90 )
plt.title('Distribution of Jobs Mentioning Degrees for US Job Postings (2023)')
plt.show()
```

```python
df_uk = df[df['job_country'] == 'United Kingdom'].copy()
```

```python
degree_mentioned_uk = df_uk['job_no_degree_mention'].value_counts()
```
